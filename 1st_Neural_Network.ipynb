## **Classification Neural Network**

```
#Keras Library for constructing Neural Network
import keras
#Numpy for Mathematical Calculation
import numpy
#Matplotlib for Visual Purpose
import matplotlib.pyplot as plt
```

## **Downloading Fashion MNIST Dataset**

```
fashion_dataset = keras.datasets.fashion_mnist
(train_images, train_labels),(test_images, test_labels) = fashion_dataset.load_data()
```

```
train_images.shape#(44 MB of memory)
```

```
train_labels[20]
```

```
class_name = ['T-shirt','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle Boot']
```

**Visualize the images**

```
plt.figure()
plt.imshow(train_images[20])
plt.show()
```

```
train_images =  train_images/255
#train_images[20]
test_images = test_images/255
```

```
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    #plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_name[train_labels[i]])
plt.show()
```

# **Building The Neural Network**

```
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28,28)))
model.add(keras.layers.Dense(128,activation='relu'))#128 Neurons
model.add(keras.layers.Dense(10))#10 Neurons
```

```
model.summary()
```

```
from keras.utils.vis_utils import plot_model
plot_model(model, to_file='MyModel.png',show_shapes=True)
```

# **Compiling Network**

```
model.compile(optimizer='adam',loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

### **Training Network**

```
model.fit(train_images,train_labels,epochs=10)
```

## Testing Model

```
test_loss, test_acc = model.evaluate(test_images,test_labels,verbose=1)
```
